```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

# We will load the datasets from our SSD
assessments = pd.read_csv('E:/Students/assessments.csv')
courses = pd.read_csv('E:/Students/courses.csv')
student_assessments = pd.read_csv('E:/Students/studentAssessment.csv')
student_info = pd.read_csv('E:/Students/studentInfo.csv')
student_registration = pd.read_csv('E:/Students/studentRegistration.csv')
student_vle = pd.read_csv('E:/Students/studentVle.csv')
vle = pd.read_csv('E:/Students/vle.csv')

# Merging the datasets
# We have merged the datasets beacause we want to build a complete dataset per student

# Aggregate StudentAssessments
student_assessment_agg = student_assessments.groupby('id_student').agg({
    'score': ['mean', 'count']
}).reset_index()
student_assessment_agg.columns = ['id_student', 'avg_score', 'assessment_attempts']

# Aggregate StudentVle
student_vle_agg = student_vle.groupby('id_student').agg({
    'sum_click': ['sum', 'mean']
}).reset_index()
student_vle_agg.columns = ['id_student', 'total_clicks', 'avg_clicks']

# Registration info
student_registration['late_registration'] = np.where(student_registration['date_registration'] > 0, 1, 0)
registration_info = student_registration[['id_student', 'late_registration']]

# Building a single student dataset
student_data = student_info[['id_student']].drop_duplicates()

# Merging all features
student_data = student_data.merge(student_assessment_agg, on='id_student', how='left')
student_data = student_data.merge(student_vle_agg, on='id_student', how='left')
student_data = student_data.merge(registration_info, on='id_student', how='left')

# Filling the missing values
student_data.fillna(0, inplace=True)

# Checking the final dataset
print(student_data.head())

# Preparing the features
features = ['avg_score', 'assessment_attempts', 'total_clicks', 'avg_clicks', 'late_registration']

X = student_data[features]

# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Training the Isolation Forest
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(X_scaled)

# Predicting the anomalies
student_data['anomaly'] = model.predict(X_scaled)

# -1 = anomaly (possible cheater), 1 = normal
cheaters = student_data[student_data['anomaly'] == -1]

print("\n--- Suspicious Students Detected ---\n")
print(cheaters[['id_student'] + features])

# Save results if needed
cheaters.to_csv('potential_cheaters.csv', index=False)
```

```python
import matplotlib.pyplot as plt
plt.scatter(student_data['avg_score'], student_data['total_clicks'], c=student_data['anomaly'])
plt.xlabel('Average Score')
plt.ylabel('Total Clicks')
plt.title('Student Behavior Clustering')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
import xgboost as xgb

# Loading our datasets
assessments = pd.read_csv('E:/Students/assessments.csv')
courses = pd.read_csv('E:/Students/courses.csv')
student_assessments = pd.read_csv('E:/Students/studentAssessment.csv')
student_info = pd.read_csv('E:/Students/studentInfo.csv')
student_registration = pd.read_csv('E:/Students/studentRegistration.csv')
student_vle = pd.read_csv('E:/Students/studentVle.csv')
vle = pd.read_csv('E:/Students/vle.csv')

# Merging and preparing the data
student_assessment_agg = student_assessments.groupby('id_student').agg({
    'score': ['mean', 'count']
}).reset_index()
student_assessment_agg.columns = ['id_student', 'avg_score', 'assessment_attempts']

student_vle_agg = student_vle.groupby('id_student').agg({
    'sum_click': ['sum', 'mean']
}).reset_index()
student_vle_agg.columns = ['id_student', 'total_clicks', 'avg_clicks']

student_registration['late_registration'] = np.where(student_registration['date_registration'] > 0, 1, 0)
registration_info = student_registration[['id_student', 'late_registration']]

student_data = student_info[['id_student']].drop_duplicates()
student_data = student_data.merge(student_assessment_agg, on='id_student', how='left')
student_data = student_data.merge(student_vle_agg, on='id_student', how='left')
student_data = student_data.merge(registration_info, on='id_student', how='left')
student_data.fillna(0, inplace=True)

# Features
features = ['avg_score', 'assessment_attempts', 'total_clicks', 'avg_clicks', 'late_registration']
X = student_data[features]

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Using isolation forest to generate pseudo labels
iso_forest = IsolationForest(contamination=0.05, random_state=42)
student_data['pseudo_label'] = iso_forest.fit_predict(X_scaled)

# Map labels: IsolationForest (-1 = anomaly) --> (1 = cheater, 0 = normal)
student_data['pseudo_label'] = student_data['pseudo_label'].map({1: 0, -1: 1})

# Training XGBoost on full data
xgb_model = xgb.XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

xgb_model.fit(X, student_data['pseudo_label'])

# Predicting
student_data['xgb_predicted_cheater'] = xgb_model.predict(X)

# Finding detected cheaters
cheaters_xgb = student_data[student_data['xgb_predicted_cheater'] == 1]

print("\n--- Students Detected as Cheaters by XGBoost ---\n")
print(cheaters_xgb[['id_student'] + features])

# Save to CSV
cheaters_xgb.to_csv('E:/Students/cheaters_detected_by_xgboost.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import xgboost as xgb

# Plot feature importance
plt.figure(figsize=(10, 6))
xgb.plot_importance(xgb_model, importance_type='gain', max_num_features=10)
plt.title('Feature Importance - XGBoost Model')
plt.show()
```

```python
import seaborn as sns

# Count predicted labels
counts = student_data['xgb_predicted_cheater'].value_counts()

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(x=counts.index, y=counts.values, palette='coolwarm')
plt.xticks([0, 1], ['Normal', 'Cheater'])
plt.ylabel('Number of Students')
plt.title('Detected Cheaters vs Normal Students')
plt.show()
```

```python
# Scatter plot: avg_score vs total_clicks
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=student_data,
    x='total_clicks', y='avg_score',
    hue='xgb_predicted_cheater',
    palette={0: 'blue', 1: 'red'}
)
plt.title('Student Behavior: Clicks vs Scores (Red = Cheaters)')
plt.xlabel('Total Clicks')
plt.ylabel('Average Score')
plt.legend(title='Cheater Prediction')
plt.show()
```
